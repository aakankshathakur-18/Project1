**Start Time Column Analysis**

```python
df.Start_Time=pd.to_datetime(df.Start_Time, format='ISO8601')
```

```python
df.Start_Time[0]
```

```python
sns.distplot(df.Start_Time.dt.hour,norm_hist=True, bins=24, kde=False)
```

```python
sns.distplot(df.Start_Time.dt.dayofweek, kde=False, norm_hist=True, bins =7)
```

**Is the distribution of accidents by hour the same on weekends as on weekend?s**

```python
sundays_start_time=df.Start_Time[df.Start_Time.dt.dayofweek==6]
sns.distplot(sundays_start_time.dt.hour, kde= False, norm_hist=True, bins=24)
```

```python
mondays_start_time=df.Start_Time[df.Start_Time.dt.dayofweek==0]
sns.distplot(mondays_start_time.dt.hour, kde= False, norm_hist=True, bins=24)
```

```python
sns.distplot(df.Start_Time.dt.month, kde=False, norm_hist=True, bins=12)
```
